```python
means = {
    "Dataset" : ["D-HAZY-MID",
                 "HTTS-SYN",
                 "I-HAZY-NTIRE-2018",
                 "O-HAZY-NTIRE-2018",
                 "Total"],
    "mean_mse" : [0,0,0,0,0],
    "mean_ssim" : [0,0,0,0,0],
}
num_dhz = 0
num_hts = 0
num_ihz = 0
num_ohz = 0

for index, row in data_hazy.iterrows():
    means["mean_mse"][4] += row["MSE"]
    means["mean_ssim"][4] += row["SSIM"]
    
    if (row["Dataset"] == "D-HAZY-MID"):
        num_dhz += 1 
        means["mean_mse"][0] += row["MSE"]
        means["mean_ssim"][0] += row["SSIM"]
    elif (row["Dataset"] == "HTTS-SYN"):
        num_hts += 1 
        means["mean_mse"][1] += row["MSE"]
        means["mean_ssim"][1] += row["SSIM"]
    elif (row["Dataset"] == "I-HAZY-NTIRE-2018"):
        num_ihz += 1
        means["mean_mse"][2] += row["MSE"]
        means["mean_ssim"][2] += row["SSIM"]
    else:
        num_ohz += 1
        means["mean_mse"][3] += row["MSE"]
        means["mean_ssim"][3] += row["SSIM"]

means["mean_mse"][0] /= num_dhz
means["mean_ssim"][0] /= num_dhz
means["mean_mse"][1] /= num_hts
means["mean_ssim"][1] /= num_hts
means["mean_mse"][2] /= num_ihz
means["mean_ssim"][2] /= num_ihz
means["mean_mse"][3] /= num_ohz
means["mean_ssim"][3] /= num_ohz
means["mean_mse"][4] /= 108
means["mean_ssim"][4] /= 108

means_hazy = pd.DataFrame(means)
means_hazy.head()
```
